```python
from spafe.utils import vis
from spafe.fbanks import bark_fbanks

# init vars
nfilts = 24
nfft = 512
fs = 16000
low_freq = 0
high_freq = 8000
scale = "constant"

# compute the bark filter banks
bark_filbanks = bark_fbanks.bark_filter_banks(nfilts=nfilts,
                                              nfft=nfft,
                                              fs=fs,
                                              low_freq=low_freq,
                                              high_freq=high_freq,
                                              scale=scale)
# plot filter banks
vis.visualize_fbanks(bark_filbanks, "Amplitude", "Frequency (Hz)")
```
