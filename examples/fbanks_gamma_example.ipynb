```python
from spafe.utils import vis
from spafe.fbanks import gammatone_fbanks

# init vars
nfilts = 24
nfft = 512
fs = 16000
low_freq = 0
high_freq = 8000
scale = "constant"

# compute the gamma filter banks
gamma_filbanks = gammatone_fbanks.gammatone_filter_banks(nfilts=nfilts,
                                                     nfft=nfft,
                                                     fs=fs,
                                                     low_freq=low_freq,
                                                     high_freq=high_freq,
                                                     scale=scale)
# plot filter banks
vis.visualize_fbanks(gamma_filbanks, "Amplitude", "Frequency (Hz)")
```
